My first attempt to implement Boots and Haraguchi model.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

from sympy.solvers import solve
from sympy import Symbol
```

```python
def eqs(x, t):
        Ndot = r*x[0] - q*(x[0]+x[1])*x[0] - beta*x[1]*x[0]
        Ydot = beta*x[1]*x[0] - capgam*x[1]
        return Ndot,Ydot
```

# Host Intrinsic Growth ($r$) against Pathogen Transmission Rate ($\beta$)

```python
m = 101

rconv = np.zeros(m)
rconc = np.zeros(m)
rsig = np.zeros(m)
beta = np.zeros (m)

for n in range(m):
    beta[n] = n/100
    rconv[n] = 1/(-2*(beta[n]+0.2)) + 2
    rconc[n] = 1/(-1*(beta[n]-1.2)) + 0.01
    rsig[n] = (2*(beta[n]-0.5))*3 - (2*(-0.5))*3
```

```python
plt.plot(beta,rconv,'c', label='Convex')
plt.plot(beta,rconc,'m', label='Concave')
plt.plot(beta,rsig, 'g', label='Sigmoidal')
plt.xlim(0,1)
plt.ylim(0,3)
plt.xlabel(r'Pathogen Transmission Rate, $\beta$', fontsize=16)
plt.ylabel(r'Host Intrinsic Growth Rate, $r$', fontsize=16)
plt.legend()
plt.show()


# Sigmoidal curve does not look right
```

# Linear Trade-off Function, r

```python
m = 51
tmax = 100000

phi=np.zeros((m,m))

N=np.zeros(m)
Y=np.zeros(m)

A = 2
B = 0.012

q = 0.003
capgam = 1

ts = np.linspace(0,10000,tmax)

for n in range(0,m):
    beta = n/50
    r = A*beta + B
    
    x0=[0.9,0.1]
    xs = odeint(eqs,x0,ts)
    N[n] = xs[tmax-1:,0]
    Y[n] = xs[tmax-1:,1]
    for i in range(0,m):
        betam = i/50
        rm = A*betam + B 
        phi[i][n]= rm - q*(N[n]+Y[n]) - betam*Y[n]
```

```python
bets=np.linspace(0,1,m)
plt.contourf(bets,bets,phi,0,colors=('w','k'))
plt.xlabel('Resident',fontsize=16)
plt.ylabel('Mutant',fontsize=16)
plt.colorbar()
```

# Sigmoidal Trade-off Function, r

```python
phi=np.zeros((m,m))

N=np.zeros(m)
Y=np.zeros(m)

A = 2
B = 0.5
C = 3

q = 0.003
capgam = 15

ts = np.linspace(0,10000,tmax)

for n in range(0,m):
    beta = n/50
    r = (A*(beta-B))*C - (A*(-B))*C
    
    x0=[0.9,0.1]
    xs = odeint(eqs,x0,ts)
    N[n] = xs[tmax-1:,0]
    Y[n] = xs[tmax-1:,1]
    for i in range(0,m):
        betam = i/50
        rm = (A*(betam-B))**C - (A*(-B))**C
        phi[i][n]= rm - q*(N[n]+Y[n]) - betam*Y[n]
```

```python
bets=np.linspace(0,1,m)
plt.contourf(bets,bets,phi,0,colors=('w','k'))
plt.xlabel('Resident',fontsize=16)
plt.ylabel('Mutant',fontsize=16)
plt.colorbar()
```
