## Find explicit equilibria

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

from sympy.solvers import solve
from sympy import Symbol
```

# Host Intrinsic Growth ($r$) against Pathogen Transmission Rate ($\beta$)

```python
m = 101

rconv = np.zeros(m)
rconc = np.zeros(m)
rsig = np.zeros(m)
beta = np.zeros (m)

for n in range(m):
    beta[n] = n/100
    rconv[n] = 1/(-2*(beta[n]+0.2)) + 2
    rconc[n] = 1/(-1*(beta[n]-1.2)) + 0.01
    rsig[n] = (2*(beta[n]-0.5))**3 - (2*(-0.5))**3
```

```python
plt.plot(beta,rconv,'c', label='Convex')
plt.plot(beta,rconc,'m', label='Concave')
plt.plot(beta,rsig, 'g', label='Sigmoidal')
plt.xlim(0,1)
plt.ylim(0,3)
plt.xlabel(r'Pathogen Transmission Rate, $\beta$', fontsize=16)
plt.ylabel(r'Host Intrinsic Growth Rate, $r$', fontsize=16)
plt.legend()
plt.show()
```

# Linear Trade-off Function, r

```python
m = 51
tmax = 100000

phi=np.zeros((m,m))

A = 2
B = 0.012

q = 0.003
capgam = 1

for n in range(0,m):
    beta = n/50
    r = A*beta + B
    
    N=capgam/(beta+0.000001)
    Y=(r-q*N)/(beta+q)
    
    for i in range(0,m):
        betam = i/50
        rm = A*betam + B 
        phi[i][n]= rm - q*(N+Y) - betam*Y
```

```python
bets=np.linspace(0,1,m)
plt.contourf(bets,bets,phi,0,colors=('w','k'))
plt.xlabel(r'Resident Strategy, $\beta$',fontsize=16)
plt.ylabel(r'Mutant Strategy, $\beta$',fontsize=16)
#plt.colorbar()
```

# Convex Trade-off Function, r

```python
phi=np.zeros((m,m))

A = -2
B = -0.2
C = 2

q = 0.05
capgam = 5

for n in range(0,m):
    beta = n/50
    r = 1/(A*(beta - B)) + C
    
    N=capgam/(beta+0.000001)
    Y=(r-q*N)/(beta+q)
    
    for i in range(0,m):
        betam = i/50
        rm = 1/(A*(betam - B)) + C 
        phi[i][n]= rm - q*(N+Y) - betam*Y
```

```python
bets=np.linspace(0,1,m)
plt.contourf(bets,bets,phi,0,colors=('w','k'))
plt.xlabel(r'Resident Strategy, $\beta$',fontsize=16)
plt.ylabel(r'Mutant Strategy, $\beta$',fontsize=16)
#plt.colorbar()
```

# Sigmoidal Trade-off Function, r

```python
phi=np.zeros((m,m))

A = 2
B = 0.5
C = 3

q = 0.003
capgam = 15

for n in range(0,m):
    beta = n/50
    r = (A*(beta-B))**C - (A*(-B))**C
    
    N=capgam/(beta+0.000001)
    Y=(r-q*N)/(beta+q)
    
    for i in range(0,m):
        betam = i/50
        rm = (A*(betam-B))**C - (A*(-B))**C
        phi[i][n]= rm - q*(N+Y) - betam*Y
```

```python
bets=np.linspace(0,1,m)
plt.contourf(bets,bets,phi,0,colors=('w','k'))
plt.xlabel(r'Resident Strategy, $\beta$',fontsize=16)
plt.ylabel(r'Mutant Strategy, $\beta$',fontsize=16)
#plt.colorbar()
```
