```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

from sympy.solvers import solve
from sympy import Symbol
```

# Figure 3a

```python
capGam=1
A = -2
B = -0.2
C = 2.5
```

```python
sym_beta = Symbol("beta")

qs = np.linspace(0,0.1,100)
ESSbet=[]

for i in qs:
    solv=solve((1/(A*(sym_beta - B))**2) + 
               ((1/(i+sym_beta))*((1/(A*(sym_beta-B)))+C-((i*capGam)/sym_beta))))
    ESSbet.append(solv)
```
